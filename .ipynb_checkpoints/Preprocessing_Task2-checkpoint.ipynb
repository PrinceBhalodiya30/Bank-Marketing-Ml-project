# Handle missing values


```python
# Mode Imputation for 'job' and 'education'
# This is safe and preserves data points.
for col in ['job', 'education']:
    mode_val = df[col].mode()[0]
    df[col] = df[col].replace('unknown', mode_val)

# Verify
print("Remaining unknown values:", (df[['job', 'education']] == 'unknown').sum().sum())
```

# Identify and handle outliers.


```python
# 1. Visualize with Boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.select_dtypes(include=['int64', 'float64']))
plt.title("Outliers Before Handling")
plt.xticks(rotation=45)
plt.show()

# 2. IQR Capping (Safe for Accuracy)
# We replace extreme values with the upper/lower whiskers instead of deleting rows.
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])

print("Outliers handled via Capping.")
```

# Encode categorical variables


```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = le.fit_transform(df[col])

print("Categorical variables encoded.")
```

# Normalize/scale numerical features.


```python
from sklearn.preprocessing import StandardScaler
# Separate target first so we don't scale it if it's classification
X = df.drop('y', axis=1)
y = df['y']

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

print("Features scaled.")
```

# EDA


```python
# Distribution of Target
sns.countplot(x=y)
plt.title("Distribution of Target Variable")
plt.show()

# Correlation Matrix
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

# Save Cleaned Data


```python
# Save the processed data to a new CSV file
# Recombine X and y for saving (optional, depending on next steps)
df_final = pd.concat([X_scaled, y], axis=1)
df_final.to_csv('bank_cleaned.csv', index=False)
print("Saved cleaned data to 'bank_cleaned.csv'")
```
